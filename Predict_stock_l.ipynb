```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print("Hello p")
```

```python
nifty50 = yf.Ticker("^NSEI")
```

```python
nifty50 = nifty50.history(period="max")
print(nifty50.columns)
print(nifty50)
```

```python
del nifty50["Dividends"]
del nifty50["Stock Splits"]
# nifty50 = nifty50.copy()
nifty50
```

```python
df = pd.DataFrame(nifty50)
df.head()
```

```python
fig,ax = plt.subplots(figsize=(10,4))
ax.plot(df.index,df["Close"], label="close price")
ax.set_xlabel("Date")
ax.set_ylabel("close price")
ax.set_title("Stock Price Over Time")
ax.legend()
plt.show()
```

```python
df["Tommorow"] = df["Close"].shift(-1)
df["Target"] = (df["Tommorow"]>df["Close"]).astype(int)
df.head()
```

```python
train = df.iloc[:-100]
test = df.iloc[-100:]
# test
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100,min_samples_split=100,random_state=1)

predictors = ["Close","Volume","Open","High","Low"]
model.fit(train[predictors],train["Target"])
```

```python
predn = model.predict(test[predictors])
predn
```

```python
from sklearn.metrics import precision_score

# when market will go up how many time it is prediction right
# It is tp/(tp+fp)
precision_score(predn,test["Target"])   
```

```python
predn = pd.Series(predn,index=test.index,name="predn")
result = pd.concat([predn,test['Target']],axis=1)
# pd.set_option('display.max_rows',None)
# pd.reset_option('display.max_row')
result
```

```python
def predict(train,test,predictors,model):
    model.fit(train[predictors],train['Target'])
    predn = model.predict(test[predictors])
    predn = pd.Series(predn,index=test.index,name="predn")
    result = pd.concat([predn,test['Target']],axis=1)
    return result
```

```python
def backtest(data,model,predictors,start=2000,step=250):
    all_predictions=[]

    for i in range(start,data.shape[0],step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train,test,predictors,model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(df,model,predictors)
predictions
```

```python
predictions["predn"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["predn"])
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = df.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    df[ratio_column] = df["Close"]/rolling_averages["Close"]

    trend_column = f"Trend_{horizon}"
    df[trend_column]=df.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column,trend_column]
```

```python
df = df.dropna()
df
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50,random_state=1)
```

```python
def predict2(train,test,predictors,model):
    model.fit(train[predictors],train['Target'])
    predn = model.predict_proba(test[predictors])[:,1]
    predn[predn >= .6] = 1
    predn[predn <= .6] = 0
    predn = pd.Series(predn,index=test.index,name="predn")
    result = pd.concat([predn,test['Target']],axis=1)
    return result
```

```python
def backtest2(data,model,predictors,start=2000,step=250):
    all_predictions=[]

    for i in range(start,data.shape[0],step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict2(train,test,predictors,model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest2(df,model,new_predictors)
predictions
```

```python
predictions["predn"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["predn"])
```
